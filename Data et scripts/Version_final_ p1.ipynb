### 1. Quelle est la note moyenne par genre des films ? Représenter graphiquement
Pour répondre à la question la première étape est de trier les données par catégorie de film. 
La seconde étape sera le calcul de la moyenne. 

Il faut d'abord gérer les données manquantes

```python
#isna scanne la table et quand la valeur est manquante vrai/faux
#Ensuite on veut compter le nombre de valeurs manquantes. 
tmdb.isna().sum() 
```

```python
#Tri du dataframe
tmdb_tri = tmdb.sort_values("genres")
#On ne va garder que les observations pour lesquels la categorie n'est pas une valeur manquante
#Il n'y a pas de valeurs manquantes dans vote_average donc pas nécéssaire de mettre une condition supplémentaire 
tmdb_tri['genres']
```

```python
#Filtrage de la dataframe pour enlever les données manquantes 
tmdb_tri = tmdb_tri[tmdb_tri["genres"].notnull()]

#Verification 
tmdb_tri.isna().sum() 
tmdb_tri['genres']
```

On a désormais bien 0 valeurs manquantes pour la variable genre

On peut désormais passer au calcul de la note moyenne 

```python
tmdb_tri.groupby("genres")["vote_average"].mean()
```

On rencontre un problème car un même film est rangé dans plusieurs catégories à la fois
Il faut faire en sorte de mettre les films autant de fois qu'ils ont de catégories dans les observations. 

```python
tmdb_tri['genres'] = tmdb_tri['genres'].str.split(', ')
tmdb_tri['genres']
#On transforme la colonne genre en mettant des listes (on met en paramètre de la méthode split la virgule)
```

str.split(sep=None, maxsplit=-1)

Return a list of the words in the string, using sep as the delimiter string. If maxsplit is given, at most maxsplit splits are done (thus, the list will have at most maxsplit+1 elements). 

If maxsplit is not specified or -1, then there is no limit on the number of splits (all possible splits are made).

https://docs.python.org/3/library/stdtypes.html#str.split

https://pandas.pydata.org/docs/reference/api/pandas.Series.str.html (pour l'utilisation de la méthode .str)

```python
tmdb_cate = tmdb_tri.explode('genres') 
#Vu qu'on a transformé les elements de la catégorie genre en listes : on peut appliquer la méthode explode qui sépare les élements d'une liste en plusieurs lignes
```

https://www.datacamp.com/tutorial/pandas-explode

```python
tmdb_cate['genres']
tmdb_cate = tmdb_cate.loc[tmdb["vote_count"] > 20]
```

**On a imposé une condition sur notre nombre de votant afin d'éviter les notes qui ne sont pas représentatives, dans le cas où le nombre de votant est très faible. On a donc considéré qu'un film est "représentatif" si le nombre de votant est supérieur à 20**

On a maintenant le tableau avec une catégorie à la fois (Un film ayant plusieurs catégories apparaît plusieurs fois).
On peut désormais calculer la moyenne par catégorie

```python
note_moyenne = tmdb_cate.groupby("genres")["vote_average"].mean()
note_moyenne
```

```python
note_moyenne.describe()
```

.unique() permet de selectionner une fois la variable, donc de supprimer les doublons (".unique() permet de selectionner une fois la variable, donc de supprimer les doublons")

```python
#graphique barplot

#On récupère les variables "genres" et "vote_moyen"

genres = tmdb_cate['genres'].unique()   
genres = sorted(genres)                # sorted() permet de trier les valeurs de notre variables considéré

vote_moyen = list(note_moyenne)

fig = plt.figure(figsize=(22,8))
sns.barplot(x=genres, y=vote_moyen, hue=genres ,palette="viridis")

plt.title("Notes moyennes par genres")
plt.xlabel("Genres")
plt.ylabel("Notes moyennes")
plt.show()
```

### 3. Analyse des facteurs influant sur la note : Analyse des facteurs influant sur la popularité : Quels critères influent sur la note d’un film ? Utiliser des graphiques pour illustrer ces relations, et mettre en avant les tendances importantes identifiées

```python
#On va d'abord catégoriser la variable genres dans tmdb_cate

#On categorise avec factorize de pandas : on assigne à chaque genre une valeur afin de construire une matrice de corrélation.
# On s'est rendu compte que ce n'était pas nécéssaire car on ne pourra pas interpréter le genre. 
cat_genres = pd.factorize(tmdb_cate["genres"]) 
cat_genresnb = pd.factorize(tmdb_cate["genres"])[0] 
tmdb_cate[["title","genres"]] 

#On creer une nouvelle variable dans tmdb_cate : cat_genres
tmdb_cate["cat_genres"] = cat_genresnb 
```

```python
#On enlève les valeurs manquantes
tmdb_cate = tmdb_cate.dropna(subset=["release_date"])
print(tmdb_cate.isna().sum())

#On reconvertit les dates en float pour pouvoir faire une heatmap
Annee = tmdb_cate["release_date"]
Annee = pd.to_datetime(Annee)
Annee = Annee.dt.strftime("%Y")        # On récupère seulement l'année de la colonne "release_date"    
tmdb_cate["release_date"] = Annee
```

```python
#On prend maintenant les variables qui nous intéresse pour construire notre matrice de corrélation. 
#On enleve également les films dont le revenu est faible. 

tmdb_cate = tmdb_cate.loc[(tmdb_cate["revenue"] > 0) & (tmdb_cate["budget"] > 0) & (tmdb_cate["runtime"] >= 30) 
&(tmdb_cate["runtime"]<= 300)]
```

Avoir un budget nul (même très faible) n'est pas cohérent puisque pour produire un film cela nécessite forcément un coût (on ne peut pas le déterminer). 
De même pour le revenu, si les recettes d'un films sont nulles cela n'est pas pertinent pour notre analyse.
De plus, on va considérer uniquement les films qui ont une durée comprise entre 30 minutes et 5 heures.

```python
# On crée la matrice de corrélation :

influ_vote = tmdb_cate[["popularity", "vote_average","runtime", "vote_count","revenue","budget","cat_genres"]]
corr_tmdb = influ_vote.corr()
display(corr_tmdb)
```

Pour créer la matrice de corrélation : ("https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html") 

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_tmdb, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation", fontsize = 20)
plt.show()


```

```python
#On prend toutes les lignes mais en prenant seulement les variables "popularity" et "vote_avearge" :
corr_finale = corr_tmdb.iloc[:,0:2]
plt.figure()
sns.heatmap(corr_finale,annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation montrant l'influence des variables considérés sur la popularité et la note", fontsize = 15)
plt.show()
```

On a crée la première matrice avec toutes les variables qui peuvent influencer la note et la popularité. Cependant on s'est rendu compte que le genre n'est pas interprétable (on a tout de même laissé pour que vous puissiez voir l'effort fourni).

Matrice de corrélation grâce à la méthode heatmap : ("https://seaborn.pydata.org/generated/seaborn.heatmap.html") 

```python
plt.figure()
sns.pairplot(influ_vote)
plt.show()
```

On a fait un pairplot afin de représenter notre matrice de corrélation par des des grpahiques qui montrent la corrélation de chaque variable deux à deux.
